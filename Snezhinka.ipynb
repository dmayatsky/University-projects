```python
import networkx as nx
```

```python
size = lambda x: x*10000
```

```python
def snezhinka(l, r):
    g=nx.balanced_tree(l,r)
    h=nx.balanced_tree(l,r)
    F=nx.Graph()
    F=nx.union(g,h, rename=('g-','h-'))
    k=dict(F.degree())
    G = []
    H = []
    for i in list(dict(k).keys()):
        if 'g' in i:
            G.append(i)
        else:
            H.append(i)
    for i, j in zip([v for v in range(len(G))],[w for w in range(len(H))]):
        if k[G[i]] == 1 and k[H[j]] == k[G[i]]:
            F.add_edge(G[i], H[j])
    return print('Диаметр:', nx.diameter(F)), print('Средняя длина пути:', nx.average_shortest_path_length(F)), nx.draw_kamada_kawai(F, with_labels=True, node_size = list(map(size, list(nx.degree_centrality(F).values()))))

    
    
```

```python
snezhinka(3,2)
```
