```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
graphs = [
          nx.read_adjlist('out.moreno_seventh_seventh', comments = '%'),
          nx.read_adjlist('out.dimacs10-football', comments = '%'),
          nx.read_adjlist('out.league-uk1-2013', comments = '%')
          ]
```

```python
from networkx.algorithms.community import greedy_modularity_communities
from networkx.algorithms.community import asyn_fluidc
```

```python
def viz_greedy_modularity_communities(G):
    dimacs = list(greedy_modularity_communities(G))
    dimacs_l = ()
    for i in dimacs:
        dimacs_l += tuple(map(str, i))
    lapl =  np.array((nx.laplacian_matrix(G, dimacs_l).todense()))
    np.fill_diagonal(lapl, 0)
    fig = px.imshow(lapl, title = 'Визуализация сообществ', contrast_rescaling = 'minmax', origin = 'upper')
    return fig.show(), print('Количество сообществ: ', len(dimacs))
```

```python
for i in graphs:
    viz_greedy_modularity_communities(i)
```

```python
def viz_asyn_fluidc(G):
    dimacs = list(asyn_fluidc(G, k = 5))
    dimacs_l = ()
    for i in dimacs:
        dimacs_l += tuple(map(str, i))
    lapl =  np.array((nx.laplacian_matrix(G, dimacs_l).todense()))
    np.fill_diagonal(lapl, 0)
    fig = px.imshow(lapl, title = 'Визуализация сообществ', contrast_rescaling = 'minmax', origin = 'upper')
    return fig.show(),  print('Количество сообществ: ', len(dimacs))
```

```python
import plotly.express as px
```

```python
for i in graphs:
    viz_asyn_fluidc(i)
```
